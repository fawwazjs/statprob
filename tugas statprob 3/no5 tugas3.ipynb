You are given the following dataset of 20 values:
65, 70, 75, 80, 85, 90, 95, 100, 70, 85, 80, 75, 90, 85, 100, 95, 85, 90, 70, 75Make the frequency distribution table, histogram, and frequency polygon.

```python
import matplotlib.pyplot as plt

# Data untuk distribusi frekuensi
class_intervals = ['65-69', '70-74', '75-79', '80-84', '85-89', '90-94', '95-99', '100-104']
frequencies = [1, 3, 3, 2, 4, 3, 2, 2]

# Mengatur posisi batang berdasarkan urutan
positions = range(len(class_intervals))

# Membuat histogram dengan grafik batang mepet tanpa celah
plt.figure(figsize=(8, 5))
plt.bar(positions, frequencies, color='skyblue', edgecolor='black', width=1)

# Mengatur label sumbu x dengan interval kelas
plt.xticks(positions, class_intervals)

# Memberikan judul dan label pada grafik
plt.title('Histogram')
plt.xlabel('Class Intervals')
plt.ylabel('Frequency')

# Menampilkan plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data untuk distribusi frekuensi
class_intervals = ['65-69', '70-74', '75-79', '80-84', '85-89', '90-94', '95-99', '100-104']
frequencies = [1, 3, 3, 2, 4, 3, 2, 2]

# Menghitung titik tengah dari tiap interval kelas
midpoints = [65,70,75,80,85,90,95,100]

# Membuat frequency polygon
plt.figure(figsize=(8, 5))
plt.plot(midpoints, frequencies, marker='o', color='red', linestyle='-', label='Frequency Polygon')

# Menambahkan area di bawah garis untuk visualisasi lebih jelas
plt.fill_between(midpoints, frequencies, color='red', alpha=0.1)

# Menambahkan judul dan label sumbu
plt.title('Frequency Polygon')
plt.xlabel('Midpoints')
plt.ylabel('Frequency')

# Menampilkan plot
plt.grid(True)  # Menambahkan grid untuk memperjelas
plt.show()
```
