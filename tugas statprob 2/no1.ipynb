Tinggi badan siswa di sebuah sekolah mengikuti distribusi normal dengan rata-rata 160 cm dan standar deviasi 7 cm. Gunakan Aturan Empiris untuk menjawab pertanyaan berikut:
o Berapa rentang tinggi badan di mana sekitar 68% siswa berada?
o Berapa rentang tinggi badan di mana sekitar 95% siswa berada?
o Berapa rentang tinggi badan di mana sekitar 99.7% siswa berada?

Diketahui: rata-rata (μ)= 160 cm
                  standar deviasi (σ)= 7cm
Ditanya: rentang tinggi badan sekitar 68%, 95%, dan 99.7% dengan menggunakan aturan empiris?
Jawab:
Rentang tinggi badan siswa sekitar 68% (μ ± σ):
μ + σ = 160 + 7 = 167 cm
μ - σ = 160 - 7 = 153 cm
Rentang tinggi badan siswa sekitar 95% (μ ± 2σ):
μ + 2σ = 160 + 2(7) = 174 cm
μ - 2σ = 160 - 2(7) = 146 cm
Rentang tinggi badan siswa sekitar 99.7% (μ ± 3σ):
μ + 3σ = 160 + 3(7) = 181 cm
μ - 3σ = 160 - 3(7) = 139 cm
Jadi rentang tinggi badan siswa sekitar 68% diantara 153cm hingga 167cm, rentang tinggi badan siswa sekitar 95% diantara 146cm hingga 174cm, dan rentang tinggi badan siswa sekitar 99.7% diantara 139cm hingga 181cm

```python
import matplotlib.pyplot as plt
import numpy as np

# fungsi untuk menkalkulasikan rentang tinggi badan menggnakan aturan empiris and plot the normal distribution
def empirical_rule_with_plot(mean, std_dev):
    # Calculating the ranges using the empirical rule
    range_68 = (mean - std_dev, mean + std_dev)
    range_95 = (mean - 2 * std_dev, mean + 2 * std_dev)
    range_99_7 = (mean - 3 * std_dev, mean + 3 * std_dev)

    # Displaying the ranges
    print(f"Rentang tinggi badan di mana sekitar 68% siswa berada: {range_68[0]} cm hingga {range_68[1]} cm.")
    print(f"Rentang tinggi badan di mana sekitar 95% siswa berada: {range_95[0]} cm hingga {range_95[1]} cm.")
    print(f"Rentang tinggi badan di mana sekitar 99.7% siswa berada: {range_99_7[0]} cm hingga {range_99_7[1]} cm.")

    # Plotting the normal distribution
    x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
    y = (1/(std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean)/std_dev)**2)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Normal Distribution', color='blue')
    
    # Shading areas for 68%, 95%, and 99.7%
    plt.fill_between(x, 0, y, where=((x >= range_68[0]) & (x <= range_68[1])), color='yellow', alpha=0.5, label='68% range')
    plt.fill_between(x, 0, y, where=((x >= range_95[0]) & (x <= range_95[1])), color='orange', alpha=0.3, label='95% range')
    plt.fill_between(x, 0, y, where=((x >= range_99_7[0]) & (x <= range_99_7[1])), color='red', alpha=0.2, label='99.7% range')

    plt.title('Distribusi Normal Tinggi Badan Siswa')
    plt.xlabel('Tinggi Badan (cm)')
    plt.ylabel('Kepadatan')
    plt.axvline(mean, color='black', linestyle='dashed', label='Rata-rata')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage with mean height = 160 cm and standard deviation = 7 cm
mean_height = 160
std_deviation = 7

empirical_rule_with_plot(mean_height, std_deviation)
```
