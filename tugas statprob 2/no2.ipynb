Dalam sebuah uji coba, waktu reaksi dari sejumlah pengemudi diukur. Diketahui bahwa waktu 
reaksi rata-rata adalah 0,8 detik dengan standar deviasi 0,1 detik. Berdasarkan Aturan Empiris, 
tentukan rentang waktu reaksi di mana:
o 68% pengemudi berada
o 95% pengemudi berada
o 99.7% pengemudi berada

Diketahui: rata-rata (μ): 0.8 detik
                  standar deviasi (σ): 0.1 detik
Ditanya: rentang waktu reaksi sekitar 68%, 95%, dan 99.7% pengemudi berada
Jawab:
Rentang waktu reaksi pengemudi berada sekitar 68% (μ ± σ):
μ + σ = 0.8 + 0.1 = 0.9 detik 
μ - σ = 0.8 - 0.1 = 0.7 detik 
Rentang waktu reaksi pengemudi berada sekitar 95% (μ ± 2σ):
μ + 2σ = 0.8 + 2(0.1) = 1 detik
μ - 2σ = 0.8 - 2(0.1) = 0.6 detik
Rentang waktu reaksi pengemudi berada sekitar 99.7% (μ ± 3σ):
μ + 3σ = 0.8 + 3(0.1) = 1.1 detik
μ - 3σ = 0.8 - 3(0.1) = 0.5 detik
Jadi rentang waktu reaksi pengemudi berada sekitar 68% diantara 0.7 detik hingga 0.9 detik, rentang waktu reaksi pengemudi berada sekitar 95% diantara 0.6 detik hingga 1 detik, dan rentang waktu reaksi pengemudi berada sekitar 99.7% diantara 0.5 detik hingga 1.1 detik.

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to calculate height ranges and plot the normal distribution
def simple_empirical_rule(mean, std_dev):
    # Calculate ranges for 68%, 95%, and 99.7%
    range_68 = (mean - std_dev, mean + std_dev)
    range_95 = (mean - 2 * std_dev, mean + 2 * std_dev)
    range_99_7 = (mean - 3 * std_dev, mean + 3 * std_dev)

    # Display ranges
    print("68% pengemudi: {range_68[0]} cm - {range_68[1]} cm")
    print("95% pengemudi {range_95[0]} cm - {range_95[1]} cm")
    print("99.7% pengemudi {range_99_7[0]} cm - {range_99_7[1]} cm")

    # Plotting
    x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
    y = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
    
    plt.plot(x, y, color='blue')
    plt.fill_between(x, 0, y, where=((x >= range_68[0]) & (x <= range_68[1])), color='yellow', alpha=0.5)
    plt.fill_between(x, 0, y, where=((x >= range_95[0]) & (x <= range_95[1])), color='orange', alpha=0.3)
    plt.fill_between(x, 0, y, where=((x >= range_99_7[0]) & (x <= range_99_7[1])), color='red', alpha=0.2)
    
    plt.axvline(mean, color='black', linestyle='--', label='Rata-rata')
    plt.title('Distribusi Reaksi Pengemudi')
    plt.xlabel('Detik')
    plt.ylabel('Kepadatan')
    plt.legend(['Distribusi Normal', '68%', '95%', '99.7%', 'Rata-rata'])
    plt.show()

# Example usage
simple_empirical_rule(0.8, 0.1)  # Mean = 160 cm, Std Dev = 7 cm
```
